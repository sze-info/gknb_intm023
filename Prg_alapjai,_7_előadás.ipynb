#Fájlkezelés

1. Fájl megnyitása, fájl objektum létrehozása (*open()*, paraméterei: a fájl neve és a megnyitás módja. r --> olvasás, read)
2. Műveletek végzése a fájlt reprezentáló objektummal (pl. a teljes tartalom szövegként történő beolvasása: *read()* metódussal)
3. Fájl lezárása (*close()* metódussal)

A fájl objektum érdekes adattagjai:
- name: a megnyitott fájl neve
- mode: megnyitás módja
- closed: lezártuk-e már

```python
# letöltés helyi gépre (forrás: http://mek.oszk.hu/00300/00384/html/)
!wget -q https://raw.githubusercontent.com/sze-info/gknb_intm023/main/adatfajlok/kisherceg20.txt
f = open('kisherceg20.txt', 'r')
print('A fájl neve: ', f.name, ', megnyitás módja: ', f.mode, ', tartalma:\n\n', f.read(), sep='')
f.close()
if f.closed:
  print('\nA fájlt lezártuk, további műveletvégzés már nem lehetséges.')
# törlés
!rm kisherceg20.txt
```

Célravezetőbb a *with* használata, mert a fájl lezárása automatikusan megtörténik még akkor is, ha valamilyen hiba szakította félbe a program futását.

```python
# letöltés helyi gépre
!wget -q https://raw.githubusercontent.com/sze-info/gknb_intm023/main/adatfajlok/kisherceg20.txt
with open('kisherceg20.txt', 'r') as f:
  print('A fájl neve: ', f.name, ', megnyitás módja: ', f.mode, ', tartalma:\n\n', f.read(), sep='')
# Nincs szükség explicit lezárásra, a *with* megteszi ezt automatikusan a behúzott blokk végén
if f.closed:
  print('\nA fájlt lezártuk, további műveletvégzés már nem lehetséges.')
# törlés
!rm kisherceg20.txt
```

Adott hosszúságú töredékek beolvasása is lehetséges a fájlból, ha a karakterek maximális mennyiségét paraméterként megadjuk a *read()*-nek. Ha elértük a fájl végét, üres stringet kapunk vissza.

```python
!wget -q https://raw.githubusercontent.com/sze-info/gknb_intm023/main/adatfajlok/kisherceg20.txt
with open('kisherceg20.txt', 'r') as f:
  while True:
    toredek = f.read(20)
    if toredek != '':
      print(toredek)
    else:
      break
!rm kisherceg20.txt
```

Ennél gyakrabban van azonban szükségünk arra, hogy a fájlból sorokat olvassunk ki. Ezt a *readline()* függvénnyel tehetjük meg (a sor tartalmazza a sorvég jelet is).

```python
!wget -q https://raw.githubusercontent.com/sze-info/gknb_intm023/main/adatfajlok/kisherceg20.txt
with open('kisherceg20.txt', 'r') as f:
  sorszam = 0
  while True:
    sor = f.readline()
    if sor != '':
      print(sorszam, ':', sor, end='')
      sorszam += 1
    else:
      break
!rm kisherceg20.txt
```

Ugyanez tömörebben, *readline()* hívás nélkül, a fájl iterálásával:

```python
!wget -q https://raw.githubusercontent.com/sze-info/gknb_intm023/main/adatfajlok/kisherceg20.txt
with open('kisherceg20.txt', 'r') as f:
  sorszam = 0
  for sor in f:
      # print(sorszam, ':', sor, end='')
      print(sorszam, ':', sor.rstrip('\n')) # a sorvég jel levágása utáni nyomtatás
      sorszam += 1
!rm kisherceg20.txt
```

A *readlines()* listába tölti a fájl összes sorát.

```python
!wget -q https://raw.githubusercontent.com/sze-info/gknb_intm023/main/adatfajlok/kisherceg20.txt
with open('kisherceg20.txt', 'r') as f:
  sorok = f.readlines()
  sorszam = 0
  for sor in sorok:
      print(sorszam, ':', sor, end='')
      sorszam += 1
!rm kisherceg20.txt
```

Szavanként is megjeleníthetjük a fájl tartalmát a string *split()* metódusát használva. (Alapértelmezetten minden fehér karakter mentén tördel, de átparaméterezhető.)

```python
!wget -q https://raw.githubusercontent.com/sze-info/gknb_intm023/main/adatfajlok/kisherceg20.txt
with open('kisherceg20.txt', 'r') as f:
  for szo in f.read().split():
    print(szo)
!rm kisherceg20.txt
```

Most, hogy már tudunk olvasni és írni, másolni is tudunk.

```python
def masol(cel: str, forras: str):
  """Fájl másolása

  Parameters
  ----------
  cel : str
    A másolat fájl azonosítója
  forras : str
    Az eredeti fájl azonosítója
  """

  with open(forras, 'r') as be:
    with open(cel, 'w') as ki:
      for sor in be:
        ki.write(sor)

def hasonlit(a: str, b: str) -> bool:
  """Két fájl tartalmának összehasonlítása

  Parameters
  ----------
  a : str
    Az egyik összehasonlítandó fájl azonosítója
  b : str
    A másik  összehasonlítandó fájl azonosítója

  Returns
  -------
  bool
    Igaz értéket szolgáltat, ha a két fájl tartalma megegyezik.
  """

  with open(a) as be_a:
    with open(b) as be_b:
      return be_a.read() == be_b.read()

masol('akos2.txt', 'akos.txt')
if hasonlit('akos.txt', 'akos2.txt'):
  print('Egyformák.')
else:
  print('Különböznek.')

!rm akos2.txt
```

## Egyszerű adatfájlok feldolgozása

Gyakran használunk szövegfájlokat mátrix jellegű adatok tárolására. Minden sor külön szövegsorba kerül, az oszlopokat pl. tabulátor karakter választja el egymástól.

```python
def kiir(mtx, fajl):
  with open(fajl, 'w') as f:
    for sor in mtx:
      f.write('\t'.join(sor) + '\n')

def beolvas(fajl):
  mtx = []
  with open(fajl, 'r') as f:
    for sor in f:
      mtx.append(sor.rstrip('\n').split('\t'))
  return mtx

m = [
     ['a1', 'a2', 'a3'],
     ['b1', 'b2', 'b3'],
     ['c1', 'c2', 'c3']
]

kiir(m, 'matrix.txt')
!cat matrix.txt
print(beolvas('matrix.txt'))
!rm matrix.txt
```

## CSV fájlok használata

Ennél egy fokkal kifinomultabb megoldás a CSV (Comma Separated Values) fájlok használata. Ez is egy struktúrált szövegfájl, de a legtöbb táblázatkezelő program ismeri, támogatja. Sajnos vannak alkalmazásfüggő eltérések, pl. a különböző nyelvi beállítások esetén a fájl tartalma eltérő lesz. Magyar Excel esetén a cellákat ; választja el egymástól, angol Excelben , karakter. A Python külön [modullal](https://docs.python.org/3/library/csv.html) rendelkezik a CSV fájlok kezeléséhez.

```python
!wget -q https://raw.githubusercontent.com/sze-info/gknb_intm023/main/adatfajlok/tablazat.csv
import csv

with open('tablazat.csv', 'r') as f:
  f.read(1) # UTF-8 formátumban mentett Excel fájl első bájtjának átlépése
  # Magyar nyelvű Excel alapbeállítása: a cellaelválasztó jel a pontosvessző, idézésre idézőjelet használ.
  olvaso = csv.reader(f, delimiter=';', quotechar='"')
  mtx = []
  for sor in olvaso:
    mtx.append(sor)
print(mtx)

!cat tablazat.csv
!rm tablazat.csv
```

A feladat legyen most az, hogy egy Excelből kimentett [táblázatot](http://rs1.sze.hu/~hatwagnf/sales.xlsx) feldolgozunk. A mentés során most nem kértünk UTF-8 formátumot, ezért az alapértelmezett karakterkódolással ([windows-1250](https://en.wikipedia.org/wiki/Windows-1250)) vannak mentve az adatok. Készítsünk függvényt, ami egy adott dátum intervallumban (OrderDate) értékesített termékek (Item) összesített számát határozza meg!

```python
# https://www.contextures.com/xlsampledata01.html#data
!wget -q https://raw.githubusercontent.com/sze-info/gknb_intm023/main/adatfajlok/sales.csv
import csv
from typing import List, Tuple # összetett type hinting esetekhez

def betolt(fajlnev : str) -> Tuple[List[str], List[List[str]]]:
  with open(fajlnev, 'r', encoding="windows-1250") as f: # karakterkódolás
    olvaso = csv.reader(f, delimiter=';', quotechar='"')
    fejlec = next(olvaso) # a következő sor olvasása
    adatok = []
    for sor in olvaso:
      adatok.append(sor)
  return fejlec, adatok

def angol2magyar(datum : str) -> str: # HH/NN/ÉÉÉÉ --> ÉÉÉÉ.HH.NN
  elemek = datum.split('/')
  return elemek[2] + "." + elemek[0] + "." + elemek[1];

def szuro(fejlec : List[str], adatok : List[List[str]], datumTol : str, datumIg : str, termek : str) -> int:
  szotar = {fej:idx for (idx, fej) in enumerate(fejlec)} # a fejléc cellák tartalmát lehessen használni a sorok indexelésére számok helyett
  datumTol = angol2magyar(datumTol)
  datumIg = angol2magyar(datumIg)
  osszeg = 0;
  for sor in adatok:
    datum = angol2magyar(sor[szotar["OrderDate"]])
    if datum>=datumTol and datum<=datumIg and termek==sor[szotar["Item"]]:
      osszeg = osszeg + int(sor[szotar["Units"]])
  return osszeg

tabla = betolt('sales.csv')
print(szuro(*tabla, "11/8/2020", "4/10/2021", "Pen"))

!rm sales.csv
```

Programból előállított adatok mentése CSV formátumba hasonlóan történik. #Adatvizualizáció [Matplotlib](https://matplotlib.org/stable/index.html)  segítségével

[Pyplot](https://matplotlib.org/stable/tutorials/introductory/usage.html): Python interfész a Matplotlibhez, Matlab-ban megszokott típusú diagramok készítéséhez. Többféle módon is elérhetők ugyanazok a szolgáltatások, de itt csak egyet mutatunk be (OO stílus), és csak az alapvető, leggyakrabban használt lehetőségekre fókuszálunk.

- [Figure](https://matplotlib.org/stable/api/figure_api.html) (fig): a teljes ábra, ami egységbe foglalja a diagramo(ka)t és azok összes komponensét (rácsos elrendezésben több diagram is elhelyezhető egymás mellett egy ábrán).
- [Axes](https://matplotlib.org/stable/api/axes_api.html) (ax): egy Figure-höz kapcsolt elem, ami az egy diagramon megjelenítő komponensek (az [Artist](https://matplotlib.org/stable/api/artist_api.html)-ok) egyike, mely egy rajzolási területet reprezentál, egyfajta vászonnak, vagy vászon darabnak tekinthető (egy Figure több diagrammal, tehát több Axes-szal is rendelkezhet). Általában kettő [Axis](https://matplotlib.org/stable/api/axis_api.html) (tengely) objektumot tartalmaz diagramonként, azokon osztásokkal, azok felirataival, tengelynevekkel, stb. - [subplots()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html): a Pyplot metódusa, mely létrehozza a Figure-t és annak egy- vagy több diagramterületét.
- [plot()](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.plot.html): *y* értékek rajzolása *x* értékek függvényében valamelyik diagramon, melyeket vonal köt össze, és/vagy jelölőkkel ([Marker](https://matplotlib.org/stable/gallery/lines_bars_and_markers/marker_reference.html)) jelzi az adatpontokat.


```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots() # Alapértelmezetten egy diagramot (ax) kapunk az ábrán (fig)
ax.plot([1, 2, 3, 4], [1, 4, 2, 3]); # A megjelenítendő x, majd y értékek listája. Ezeket alapértelmezetten vonalak kötik össze.
# A tengelyek beosztása automatikusan történik.
```

Kérhetünk feliratot
- az egész diagramhoz ([set_title()](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_title.html)), pl. "Függvények"
- az x tengelyre ([set_xlabel()](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_xlabel.html)), pl. "x"
- az y tengelyre ([set_ylabel()](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_ylabel.html)), pl. "f(x)"

Jelmagyarázatot a [legend()](https://matplotlib.org/stable/tutorials/intermediate/legend_guide.html) metódus jelenít meg. Olyan helyet keres, ahol a legkevésbé takarja az adatokat. Tartalma a plot() hívásban megadott *label* paraméter értékétől függ.

Több görbét is felrajzolhatunk egy diagramra. Ekkor automatikusan más-más beállításokat használ a matplotlib, hogy a görbék megkülönböztethetők legyenek.

```python
import matplotlib.pyplot as plt
import math
fig, ax = plt.subplots()
ax.set_title("Függvények") # Diagram felirata
ax.set_xlabel("x")         # Az x tengely felirata
ax.set_ylabel("f(x)")      # Az y tengely felirata
PRECISION = 50             # Pontosság; ennyi adatpontot generálunk, 
                           # hogy sima görbéket kapjunk
# A range() csak egész számokkal tud dolgozni, ezért a vizsgált tartomány 
# [0, 2pi] PRECISION darab xi értékének előállítására közvetetten használjuk.                           
x = [(x/(PRECISION-1))*(2.*math.pi) for x in range(PRECISION)]
sin_y = [math.sin(xi) for xi in x] # Minden xi értékhez meghatározza a sin() fv. értékét
cos_y = [math.cos(xi) for xi in x] # Ua., csak cos()-szal
ax.plot(x, sin_y, label="sin()")   # 'label' értéke lesz a görbe jelmagyarázatban feltüntetett szövege
ax.plot(x, cos_y, label="cos()")
ax.legend();                       # Jelmagyarázat megjelenítése
```

Ha a görbéknek automatikusan választott színek valamiért nem megfelelőek (pl. mert monokróm lézernyomtatón kinyomtatva nem különböztethetőek meg jól az eltérő színű görbék), akkor beállíthatunk általunk meghatározott színeket, vonalstílusokat, vastagságokat.

Még a jelmagyarázat helye is megváltoztatható, ha pl. azt szeretnénk elérni, hogy több ábránk egységes megjelenést kapjon.

- [Vonalstílusok](https://matplotlib.org/3.5.1/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D.set_linestyle) (linestyle)
- [Színek](https://matplotlib.org/3.5.1/tutorials/colors/colors.html) (color)
- [Vastagság](https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D.set_linewidth) (linewidth)
- [Jelmagyarázat pozíciója](https://matplotlib.org/3.5.1/api/_as_gen/matplotlib.pyplot.legend.html)

```python
import matplotlib.pyplot as plt
import math
fig, ax = plt.subplots()
ax.set_title("Függvények")
ax.set_xlabel("x")
ax.set_ylabel("f(x)")
PRECISION = 50
x = [(x/(PRECISION-1))*(2*math.pi) for x in range(PRECISION)]
sin_y = [math.sin(xi) for xi in x]
cos_y = [math.cos(xi) for xi in x]
# Piros szín (színkonstanssal megadva), 1.5 pont vonalvastagság, szaggatott vonalstílus
ax.plot(x, sin_y, label="sin()", color="red", linewidth=1.5, linestyle="--")
# Fekete szín (RGB összetevők intenzitásával megadva), 2.5 pont vonalvastagság, pontvonal (kulcsszóval)
ax.plot(x, cos_y, label="cos()", color=(0, 0, 0), linewidth=2.5, linestyle="dotted")
ax.legend(loc="upper right"); # Jelmagyarázat a jobb felső sarokban
```

Az adatpontokat külön meg lehet jelölni, hangsúlyozandó, hogy valójában csak ezek az adatok állnak rendelkezésre, az összekötő vonalak mentén leolvasható értékek csak feltételezések. (Az előző ábrákon olyan sok volt belőlük, hogy nem lett volna értelme külön jelölni őket.) Az ábrán megjelenítendő [jelölők](https://matplotlib.org/3.5.1/api/markers_api.html) tulajdonságai a *marker* paraméteren keresztül állíthatók be.

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [1, 4, 2, 3], marker="D"); # 'D' mint diamond, azaz gyémánt
```

Lehet, hogy az összekötő vonalakra nincs is szükség (mert pl. azt a hamis illúziót keltenék, hogy a vizsgált függvény folytonos). Ilyenkor elegendő csak a jelölőket feltüntetni. Ezt legkönnyebben a plot() *fmt* paraméterével tehetjük meg, mely egy formázott karakterlánc, és a következő, elhagyható összetevőkből áll:

- Jelölő, pl. '.' a ponthoz, 'o' a körhöz, '*' a csillaghoz, stb.

Ha csak jelölőket szeretnénk a diagramon feltüntetni, valószínűleg kifejezőbb a [scatter()](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.scatter.html) diagramtípus (pontdiagram, buborék diagram) használata. A megjelenés testre szabható:

- marker: a jelölő típusa (pl. 'o' esetén kör)
- s: a jelölő mérete (pont négyzetében megadva)
- facecolor: kitöltőszín, pl. "red" esetén vörös

Az ábra méretét a subplots()-nak átadott *figsize* paraméter tuple-lel lehet beállítani a kívántra (az ábra szélessége és magassága inch-ben, alapértelmezetten 6.4 és 4.8). Ennek nyilván csak a felbontással (*dpi*) együtt van értelme, ami alapértelmezetten 100 értékű.

```python
import matplotlib.pyplot as plt
```
```python
# egy sorban két diagram a közös ábrán
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(16, 9))
N = 25
# Az x, és ennek következtében az y értékek szándékosan nem 0-val kezdődnek, 
# hiszen azt nem lehetne a logaritmikus skálán ábrázolni
x = [x for x in range(1, N)]
y = [x**2 for x in range(1, N)] 
ax1.grid(True) # háttér rácsozásának bekapcsolása
ax1.plot(x, y)
ax2.grid(True)
ax2.set_yscale('log') # Az y tengely logaritmikus osztású
ax2.plot(x, y);
```

A diagramokon elhelyezett különféle [szövegek formázhatóak](https://matplotlib.org/3.5.1/tutorials/text/text_props.html). Pl. a diagram felirata lehet 24 pontos betűkkel szedett (*fontsize*), kék színű (*color*), talp nélküli karakterkészlettel írt (*family*) és dőlt betűs (*style*).

Nem csak a már megismert feliratok formázhatók ilyen módon, de a diagramon tetszőleges helyre helyezhetünk [szöveget](https://matplotlib.org/3.5.1/api/_as_gen/matplotlib.axes.Axes.text.html), amit ugyanígy megformázhatunk.

A szöveget megadhatjuk [(La)TeX formázással](https://matplotlib.org/3.5.1/tutorials/text/mathtext.html), amivel könnyű görög betűket és egyenleteket, speciális jelöléseket (pl. törtek, gyökvonás, integrálás), indexelést, stb. megadni.

Arra is van mód, hogy [feliratokat készítsünk egy-egy meghatározott adatponthoz](https://matplotlib.org/3.5.1/gallery/text_labels_and_annotations/annotation_demo.html).

```python
import matplotlib.pyplot as plt, random, math

# Síkbeli pont origótól való távolsága
def distance(x, y):
  return math.sqrt(x*x + y*y);

fig, ax = plt.subplots(figsize=(12, 8))
N = 1000
x = [random.gauss(0, 1) for xi in range(N)]
y = [random.gauss(0, 1) for xi in range(N)]
max_dist = -1 # Az eddig fellelt legnagyobb távolság
for (xi, yi) in zip(x, y):
  d = distance(xi, yi)
  if d > max_dist:
    max_dist = d
    dist_x, dist_y = xi, yi # Alegtávolabb lévő pont helye
ax.axis([-6, +6, -4, +4]) # A tengelyek mentén ábrázolható legkisebb és legnagyobb értékek
# Betméret, írásszín, karaktercsalád, stílus (álló, dőlt) beállítása
ax.set_title("Normális (Gauss) eloszlás", fontsize="24", color="blue", family="sans-serif", style="italic")
# Az első két paraméter a szöveget befoglaló téglalap alakú terület egyik sarkának pozíciója
# az adatok koordinátarendszerében megadva. Nálunk az x=6 érték a diagram jobb szélét,
# az y=4 pedig a tetejét jelenti, tehát a viszonyítási pont a diagram jobb felső sarka.
# Ez a pont a téglalap jobb (horizontalalignment="right") felső (verticalalignment="top") sarkának helye.
# A felirat (La)TeX formázással készült a görög betűk megjelenítése miatt. 
ax.text(6, 4, r'$\mu=0,\ \sigma=1$', fontsize="14", horizontalalignment="right", verticalalignment="top")
# Egy adatpont, most az origótól legtávolabbi feliratozása. *xy* a megjelölendő pont helye,
# *xytext* a felirat helye a diagramon, az *arrowprops* pedig a kettő közötti nyíl tulajdonságait
# adja meg: *facecolor*  a színét, *shrink* pedig a rövidítés mértékét (hogy ne érjen hozzá a szöveghez 
# és a mutatott ponthoz).
ax.annotate("Legtávolabbi elem", xy=(dist_x, dist_y), xytext=(-5, -3), arrowprops={"facecolor": "black", "shrink": 0.05})
ax.grid(True)
ax.scatter(x, y);
```

Időnként a mérési [hibákat](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.errorbar.html#matplotlib.axes.Axes.errorbar), az adatok bizonytalanságát is ki szeretnénk fejezni.
```python
import matplotlib.pyplot as plt
import random

mu = 100  # eloszlás átlaga
sigma = 15  # eloszlás szórása
x = [random.gauss(mu, sigma) for x in range(400)]
bins = 10 # Az oszlopok (mérési adatok tartományainak) száma
fig, ax = plt.subplots()
ax.set_xlabel('Smarts')
ax.set_ylabel('Probability density')
ax.set_title(r'Histogram of IQ: $\mu=' + str(mu) +'$, $\sigma=' + str(sigma) +'$')
# density=True esetén nem a nyers darabszámokat mutatják az oszlopok, hanem a 
# valószínűségi sűrűségfüggvény értékét, azaz a hisztogram területe 1.
ax.hist(x, bins, density=True);
```

Kétváltozós függvények értékét síkon megjeleníteni nem könnyű, de egy [hőtérkép](https://matplotlib.org/stable/gallery/images_contours_and_fields/image_annotated_heatmap.html) (amin a függvény értéke egy ábra valamely területének, vagy egyetlen pontjának színével adott) segíthet. Az alkalmazott színek jelentését egy [sávról](https://matplotlib.org/stable/api/colorbar_api.html) lehet leolvasni.

A [Rastrigin](https://en.wikipedia.org/wiki/Rastrigin_function) függvény gyakran alkalmazott tesztfüggvény optimalizációs algoritmusok működésének ellenőrzésére, mérésére. Folytonos függvény, ami tetszőleges számú független változóra általánosítható, mi itt a kétváltozós változatát fogjuk használni. Globális minimumát a független változók 0 értékénél fogja felvenni.

```python
import matplotlib.pyplot as plt, math

# n-változós Rastrigin fv.
def rastrigin(*x):
  A = 10
  n = len(x)
  sum = A*2
  for xi in x:
    sum += xi**2-A*math.cos(2*math.pi*xi)
  return sum

PRECISION = 100
RANGE = 5.12
adatok = [[rastrigin(RANGE-2*RANGE*x/(PRECISION-1), RANGE-2*RANGE*y/(PRECISION-1)) for y in range(PRECISION)] for x in range(PRECISION)]
fig, ax = plt.subplots(figsize=(10, 10))
# A tengelyek mentén a mátrix formában adott adatok indexei jelennének meg, ezt
# lecseréljük az extent paraméter segítségével. Most a két független változó
# értékeit mutatja a diagram.
pos = ax.imshow(adatok, extent=[-RANGE, RANGE, -RANGE, RANGE])
# A színsáv méretét lehet csökkenteni a shrink paraméterrel.
fig.colorbar(pos, ax=ax, shrink=0.75);
```
